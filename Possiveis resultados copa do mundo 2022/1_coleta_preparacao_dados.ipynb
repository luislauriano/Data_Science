#### Calculating Elo Rating

Elo Rating é um método para calcular os níveis relativos de habilidade dos jogadores em jogos de soma zero, como o xadrez. É nomeado após seu criador Arpad Elo, um professor de física húngaro-americano.

Isso pode ser útil para avaliarmos as seleções mais fortes da copa de 2022. Será construído com base no histórico de partidas internacionais da FIFA até 2022.

```python
def get_result(df):
    if df['home_score'] > df['away_score']:
        return 1
    elif df['home_score'] == df['away_score']:
        return 0.5
    else:
        return 0
df_partidas_internacionais['result'] = df_partidas_internacionais.apply(get_result, axis=1)
```

```python
teams_df = pd.unique(df_partidas_internacionais[['home_team', 'away_team']].values.ravel('K'))
teams_df = pd.DataFrame({'team': teams_df})
teams_df['elo'] = 1500
```

```python
teams_df = teams_df.set_index('team')
```

```python
def expected(A, B):
    """
    Calculate expected score of A in a match against B
    :param A: Elo rating for player A
    :param B: Elo rating for player B
    """
    return 1 / (1 + 10 ** ((B - A) / 400))


def elo(old, exp, score, k=30):
    """
    Calculate the new Elo rating for a player
    :param old: The previous Elo rating
    :param exp: The expected score for this match
    :param score: The actual score for this match
    :param k: The k-factor for Elo (default: 30)
    """
    return old + k * (score - exp)
```

```python
#teams_df.set_index('team', inplace=True)

elo_ratings_home, elo_ratings_away = [], []

for index, row in df_partidas_internacionais.iterrows():
    elo_home = teams_df['elo'].loc[row['home_team']]
    elo_away = teams_df['elo'].loc[row['away_team']]
    
    new_elo_home = elo(elo_home, expected(elo_home,elo_away),row['result'])
    new_elo_away = elo(elo_away, expected(elo_away,elo_home),(row['result']-1)*-1)

    teams_df['elo'].loc[row['home_team']] = new_elo_home
    teams_df['elo'].loc[row['away_team']] = new_elo_away
    
    elo_ratings_home.append(new_elo_home)
    elo_ratings_away.append(new_elo_away)
    
df_partidas_internacionais['home_elo'] = elo_ratings_home
df_partidas_internacionais['away_elo'] = elo_ratings_away
```

```python
# incluir os países participantes como um vetor quente
df_partidas_internacionais['wc_participant'] = df_partidas_internacionais['home_team'] * df_partidas_internacionais['home_team'].isin(df_copa_2022.index.tolist())
df_partidas_internacionais['wc_participant'] = df_partidas_internacionais['wc_participant'].replace({'':'Other'})
df_partidas_internacionais = df_partidas_internacionais.join(pd.get_dummies(df_partidas_internacionais['wc_participant']))
```
